## Correlation and interaction
In the following cells, we try to find out if there is a correlation between some of the weather features and the produced KW of energy.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from statsmodels.regression import linear_model

from preprocessors.paths import (PATH_TO_PLANT_A,
                                 PATH_TO_PLANT_B,
                                 PATH_TO_PLANT_C,
                                 PATH_TO_WEATHER)
from preprocessors.plant_preprocessor import PlantPreprocessor
from preprocessors.weather_preprocessor import WeatherPreprocessor
```

```python
# We only use the data from the power plant A in this calculation. For further investigation, 
# you might also want to include power plants B and C.

data_power_plant_a = PlantPreprocessor(PATH_TO_PLANT_A).df_indexed_utc
data_weather = WeatherPreprocessor(PATH_TO_WEATHER).df_indexed_utc
```

```python
print('data_power_plant_a columns:', list(data_power_plant_a.columns),
      'data_weather columns:', list(data_weather.columns), 
      sep='\n')
```

```python
data_power_plant_a.head()
```

```python
data_weather.head()
```

```python
data_power_plant_a.shape
```

```python
data_weather.shape
```

```python
# To be able to create a correlation matrix, we need to add the data of the generated KW to the dataframe 
# with the weather data. The parameter join_axes is necessary, as otherwise the fitting of the linear model will 
# fail due to an index-error.

df_all = pd.concat([data_weather, data_power_plant_a['generation_kw']], axis=1, join_axes=[data_weather.index])
df_all.head()
```

```python
# To simply look for correlations, we do not need the timestamp here. We just need to make sure that 
# the data is joined in the right way as we did in the previous step.

df_all = df_all.reset_index()
del df_all['timestamp_utc']

df_all.head()
```

```python
df_all = df_all.dropna()
df_all.shape
```

```python
# In the following matrix you will see which features correlate with each other.
# A bright green indicates a positive correlation and a dark green a negative one.

plt.matshow(df_all.corr())
plt.xticks(range(df_all.shape[1]), df_all.columns, fontsize=14, rotation=90)
plt.yticks(range(df_all.shape[1]), df_all.columns, fontsize=14)
plt.show()
```

```python
# The next matrix show the correlation values for each pair of features. A value close to 1 (or -1) 
# indicates o high correlation.

corr = df_all.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
# As we can see in the matrix above, the radiation (surface and toa) and the temperature seem to be 
# the features which influence the generated KW the most. Although this is not surprising at all, we want 
# to check if there are some interactions between features. For this we take the radiation_surface, 
# the temperature and the air_density.

X = df_all[['radiation_surface', 'temperature', 'air_density']]
y = df_all['generation_kw']
print(X.shape)
print(y.shape)
```

```python
X.head()
```

```python
y.head()
```

```python
# Here we need to consider all possible combinations of interactions of the three features.

x_interaction = PolynomialFeatures(2, interaction_only=True, include_bias=False).fit_transform(X)

interaction_df = pd.DataFrame(x_interaction, columns = ['radiation_surface','temperature','air_density',
                                                        'radiation_surface:temperature','radiation_surface:air_density',
                                                        'temperature:air_density'])
```

```python
# For the fitting of the linear model, we use the ordinary least squares method (OLS).

interaction_model = linear_model.OLS(y, interaction_df).fit()
```

```python
interaction_model.pvalues[interaction_model.pvalues < 0.05]
```

### Interpretation
The above listed features and interactions are all considered to be statistically significant for predicting the generated KW of energy. As we can see there seem to be an interaction between radiation_surface and temperature as well as between radiation_surface and air_density. Looking at the p-values, the combination of radiation_surface and temperature seem to be the most important feature.
For this calculation we used all data including the timesteps where no energy was produced (e.g. at night). In the following cells we will fit the same model again only using the observations where actually some energy was produced.

```python
df_all_kw = df_all[df_all.generation_kw != 0.0]
df_all_kw.head()
```

```python
corr2 = df_all_kw.corr()
corr2.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
# There does not seem to be a huge difference between the two correlation matrices. Nevertheless, 
# let's see how the linear model will look like. For this model, we will only have more or less have 
# of the observations.

X = df_all_kw[['radiation_surface', 'temperature', 'air_density']]
y = df_all_kw['generation_kw']
print(X.shape)
print(y.shape)
```

```python
# The data of the y axes needs to be explicitely transformed to a list in this case.
# Otherwise the fitting will fail due to an index-error. We do not know the exact reason for this, as we 
# ensured the correct concatination of the data further up in the code.

y = list(y)

x_interaction = PolynomialFeatures(2, interaction_only=True, include_bias=False).fit_transform(X)

interaction_df = pd.DataFrame(x_interaction, columns = ['radiation_surface','temperature','air_density',
                                                        'radiation_surface:temperature','radiation_surface:air_density',
                                                        'temperature:air_density'])
```

```python
interaction_model = linear_model.OLS(y, interaction_df).fit()
```

```python
interaction_model.pvalues[interaction_model.pvalues < 0.05]
```

Although the correlation matrices were only slightly different, there seem to be more features considered to be statistically significant in the linear model. But as in the first model the most relevant feature is the interaction between radiation_surface and temperature. Nevertheless the air_density does not seem unimportant and might should be consired for further analysis.
